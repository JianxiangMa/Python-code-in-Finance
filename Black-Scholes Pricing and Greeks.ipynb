```python
import numpy as np
from scipy.stats import norm
```

```python
def bs_call(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    d_2 = d_1 - sigma * np.sqrt(T)
    return S * norm.cdf(d_1) - K * np.exp(-r * T) * norm.cdf(d_2)
```

```python
def bs_put(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    d_2 = d_1 - sigma * np.sqrt(T)
    return K * np.exp(-r * T) * norm.cdf(-d_2) - S * norm.cdf(-d_1)
```

```python
def bs_call_delta(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    return norm.cdf(d_1)
```

```python
def bs_put_delta(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    return norm.cdf(d_1) - 1
```

```python
def bs_gamma(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    return norm.pdf(d_1) / (S * sigma * np.sqrt(T))
```

```python
def bs_vega(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    return S * np.sqrt(T) * norm.pdf(d_1)
```

```python
def bs_call_rho(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    d_2 = d_1 - sigma * np.sqrt(T)
    return K * T * np.exp(-r * T) * norm.cdf(d_2)
```

```python
def bs_put_rho(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    d_2 = d_1 - sigma * np.sqrt(T)
    return -K * T * np.exp(-r * T) * norm.cdf(d_2)
```

```python
def bs_call_theta(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    d_2 = d_1 - sigma * np.sqrt(T)
    return -(S * norm.pdf(d_1) * sigma) / (2 * np.sqrt(T)) - r * K * np.exp(-r * T) * norm.cdf(d_2)
```

```python
def bs_put_theta(S,K,T,r,sigma):
    d_1 = (np.log(S / K) + (r + 0.5 * (sigma ** 2)) * T) / (sigma * np.sqrt(T))
    d_2 = d_1 - sigma * np.sqrt(T)
    return  -(S * norm.pdf(d_1) * sigma) / (2 * np.sqrt(T)) + r * K * np.exp(-r * T) * norm.cdf(-d_2)
```
